# The Structure of a Script/Program

## Computer Programs in Brief

Python code, like most programming languages, is organized as follows:

 - An "entry point" is a "file" where execution starts
 - That initial file contains a sequence of instructions for the "interpreter"
 - The "instructions" might include:
   - (Functions) Groups of instructions that may be used later (often multiple times)
   - (Flow Control) Instructions that coordinate whether to do something, or something else (e.g.: "if remaining tries is more then zero, then do whatever, else display an error message" or "for each item in a list, do something to that item")
   - "Other things to do" (e.g.: storing data, displaying data, manipulating data, ...)


A good way to visualize the execution of code is using three items:

 - a flow chart
 - a white board
 - a little arrow pointing to the a step in the flow chart
 
The white board is used to store information in an organized fashion. Sequences of items and individual items are named and recorded.

The flow chart (along with a pointer to the current step) tells us where we are in a program, and where to go next. Blocks might consult or modify information on the white board.

Those who have been somewhat exposed to computer architecture might see the white board being like a "data segment", the entire flow chart being like a "code segment", and the pointer to where we are being like an "instruction pointer".

## Formatting Text

Python has a format() method on strings which can help to sub in text which we need.

````python
name = "Bob"
hello_text = "Hello, {}".format(name)
print(hello_text)

fruit = "apples"
print("{} likes to eat {}.".format(name, fruit)) # sub in by order

print("{name} likes to eat {fruit}.".format(name=name, fruit=fruit)) # named way

pi = 3.14159265359
print("Pi to 2 dp: {}".format(round(pi, 2)))
print("Pi to 3 dp: {}".format(round(pi, 3)))
print("Pi to 4 dp: {}".format(round(pi, 4)))
````

## Code Blocks

Code blocks are sequences of instructions.

Python uses the whitespace (tabs or spaces) to define code blocks.

````python
print("Hello")                # in Top Level Block
print()                       # in Top Level Block
for number in range(1, 6):    # in Top Level Block
    print()                   # Block scoped to "for loop"; also in Top Level Block
    print(number)             # Block scoped to "for loop"; also in Top Level Block
    if (number % 2 == 0):     # Block scoped to "for loop"; also in Top Level Block
        print("... is even")  # In Block for "If X"; also in Block scoped to "for loop" and Top Level Block
        print()               # In Block for "If X"; also in Block scoped to "for loop" and Top Level Block
    else:                     # Block scoped to "for loop"; also in Top Level Block
        print("... is odd")   # In Block for "Else"; also in Block scoped to "for loop" and Top Level Block
        print()               # In Block for "Else"; also in Block scoped to "for loop" and Top Level Block
                              # in Top Level Block
print("All done")             # in Top Level Block
````

## Iteration ("Loops")

Any collection can be iterated.

````python
for x in range(7):
    print(x)
    print('\tSquared:', x*x)
````

````python
for x in range(1, 7):
    print(x) 
````

````python
for x in range(10):
    if x == 1:
        continue  # finishes the current iteration and continues with the next
    if x > 5:
        break     # ends the entire loop
    print(x)
````

````python
x = 0 
while (x < 10):
    print(x)
    x += 1 # x = x + 1
````

## Dictionaries

Dictionaries are associative arrays. An element (a key) is associated with another (a value).

````python
captains = {} 

captains["Enterprise"] = "Kirk"
captains["Enterprise D"] = "Picard"
captains["Deep Space Nine"] = "Sisko"
captains["Voyager"] = "Janeway"

captains
````

Retrieving elements by key...

````python
captains["Voyager"]
````

````python
try:
    captains["no-such-ship"]
except Exception as e:  # also an exception handling example
    print(e.__class__, e.__doc__, e)
````

If key does not exist...

````python
print(captains.get("Enterprise"))
print(captains.get("no-such-ship"))
````

Looping through keys and values

````python
for ship, captain in captains.items():  # by keys & values (items)
    print("{ship}: {captain}".format(ship=ship, captain=captain))
````

# Functions

The first line of a function definition contains:
 - `def`
 - the name of the function (say: `function_name`)
 - the arguments the function accepts in parentheses (say: `argument1`, `argument2`, and `argument3`)
 - a `:`

... it is then followed by the "function body".

Like so:

```
def function_name(argument1, argument2, argument3):
    # function body in this block
    # do stuff
    return result  # obviously this won't run
```

A function body is a block containing the instructions to be executed. These might include `return` statements, that terminate the function and "return" (send back) to the caller the accompanying "return value".

More examples below.


If a function does not end with a return statement, the compiler inserts a `return None`. (This can be verified by examining byte code using the `dis` module.)

Functions are good for recording "standard operating procedures" that are to be executed multiple times in "standard ways" (e.g.: having a consistent dependency on some input data).

````python
def square_something(x):
    return x * x

print(square_something(2))
print(square_something(12))
````

````python
def relu(x):
    if x < 0:
        return 0
    else:
        return x

print(relu(-2))
print(relu(2))
print(relu(12))
````

## Functions a "First Class Citizens"

You can pass functions around as parameters and return them as output... just like any other type of data.

````python
def cube_something(x):
    return x * x * x

print(cube_something(2))
````

````python
def apply_function(f, x):
    return f(x)

apply_function(cube_something, 3)
````

## Anonymous Functions

`lambda`'s are anonymous functions

Most common use: simple inline functions passed as arguments to functions

````python
apply_function(lambda x: x * x * x, 3)
````
